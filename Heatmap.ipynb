```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm, rcParams
import seaborn as sns
import numpy as np
sns.set()
```

```python
dataset = pd.read_csv(r'C:\Users\User\Downloads\AM_IM_SS_13.csv')
data = dataset.copy()

```

```python
corr = dataset.corr()
corr
```

```python
fig=plt.gcf()

sns.set(font="Times New Roman", font_scale=1.2)
sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap='Blues')
```

```python
columns = np.full((corr.shape[0]), True, dtype=bool)

for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i, j] >= 0.6:
            if columns[j]:
                columns[j] = False
                
columns
```
